```python
import numpy as np
import matplotlib.pyplot as plt
from functools import reduce

```

\begin{eqnarray}
E &=& - J \sum_{\langle ij \rangle} \cos(\theta_{i}-\theta_{j})\\
E^{u} &=& - J \sum_{\langle ij \rangle} \cos(\theta_{i}^{u}-\theta_{j}^{u})\\
E^{v} &=& - J \sum_{\langle ij \rangle} \cos(\theta_{i}^{v}-\theta_{j}^{v}) \\
\Delta E &=& E^{v} - E^{u} \\
\end{eqnarray} 
Due to local nature of algorithm only change at site $i$ is $\theta^{u}_{i} \rightarrow \theta^{v}_{i}$
\begin{eqnarray}
E^{u} &=& - J \sum_{\langle ij \rangle} \cos(\theta_{i}^{u}-\theta_{j}^{u})\\
E^{v} &=& - J \sum_{\langle ij \rangle} \cos(\theta_{i}^{v}-\theta_{j}^{u}) \\
\Delta E &=& J \sum_{\langle ij \rangle}(\cos(\theta_{i}^{u}-\theta_{j}^{u}) - \cos(\theta_{i}^{v}-\theta_{j}^{u})) \\
\Delta E &=& J \sum_{\langle ij \rangle}(\cos(\theta_{i}^{u})\cos(\theta_{j}^{u}) + 
\sin(\theta_{i}^{u})\sin(\theta_{j}^{u}) )
\end{eqnarray} 

### Slow cooling
1. **`How to perform slow cooling from high T to low T phase`**.
1. Perform a sudden quench from $T \rightarrow \infty$ to $T_{0}$.
    * One can set $T_{0}$ to $2 T_{c}$.
    * We thermalize the system at $T_{0}$ ( perform lot of MCS (10^{5}).
    * Now change the temperature and make the measurement at the new temperature (T) `no thermalization required`. 
    
1. Perform a infinetly random quench.
    * Use an initial lattice configuration as an initial state.
    * Now perform monte carlo for temperature (T) `without any equilibration`.
1. We cool system by changing the temperature linearly according to following criterion.
    * $T(t) = \frac{T_{0}}{2}(1-\frac{t}{\tau_{q}})$
    * At, $t \rightarrow -\tau_{q}, T(t) = T_{0} = 2T_{c}$
    * At, $t = 0, T(t) = \frac{T_{0}}{2} = T_{c}$.
    * At, $t \rightarrow \tau_{q}, T(t) = 0$.
    * One can fix $\tau_{q}$ and use different values of $\tau_{q}$.

```python
### Fixed parameters for a xy model
L = 16
n_sites = L* L
J = 1.0

### Simulation parameters
n_equil = 10000
n_meas = 1

### setting up the random number seed
np.random.seed(1)
```

```python
class Spin():
    ### initializing the object for a given system size L and system parameters
    def __init__(self,L=L,n_equil=n_equil,n_meas=n_meas,J=J):
        self.L = L
        self.Equil = n_equil
        self.n_meas = n_meas
        self.J = J
        self.enr = None
        self.mag = None
        self.enr_list = []
        self.mag_list = []
        self.norm_thetaconf = []
        self.winding_num=[]
        
    ### calculating the number of sites present in the system
    def calc_n_sites(self):
        self.n_sites = (self.L)**2
        self.sites = np.arange(0,self.n_sites,1) 
        
        
    ### calculating neighbours of the spins
    def neighbour(self):
        tempr = [] ; templ = []
        tempu = [] ; tempd = []
        for i in range(self.n_sites):
            xi = i%(self.L)
            yi = i//(self.L)
            tempr.append(( (xi+1)%self.L+ yi*self.L))
            templ.append(( (xi-1+self.L)%self.L+ yi*self.L))
            tempu.append((xi+((yi+1)%self.L)*self.L ))
            tempd.append((xi + ((yi-1+self.L)%self.L)*self.L))
        self.right = tempr
        self.left = templ
        self.up = tempu
        self.down = tempd   
    
    ### creating arrays of spins (theta) 
    ### initialize the lattice with possible random values
    def theta_init(self):
        self.site_theta = np.random.uniform(-np.pi,np.pi,self.n_sites)
    
    
    ### single flip algorithm 
    ### the choice of random field is passed as an argument (field)
    ### the sequence of random spins that are choosen for the flipping are
    ### passed as an array rand_sites
    ### theta is change to field if the energy is minimized else with a probability

    def montecarlo_sweep_single_flip(self,temp,field,rand_sites):
        for site in rand_sites:
            ## calculating energy cost due neighbouring spin configurations
           
            rs_i = self.right[site]
            ls_i = self.left[site]
            us_i = self.up[site]
            ds_i = self.down[site]
            
            ## theta_i - theta_j for initial configurations
            
            term_1u = self.site_theta[site] - self.site_theta[rs_i]
            term_2u = self.site_theta[site] - self.site_theta[ls_i]
            term_3u = self.site_theta[site] - self.site_theta[us_i]
            term_4u = self.site_theta[site] - self.site_theta[ds_i]
            
            
            ## theta_i - theta_j for final configurations
            
            term_1v = field[site] - self.site_theta[rs_i]
            term_2v = field[site] - self.site_theta[ls_i]
            term_3v = field[site] - self.site_theta[us_i]
            term_4v = field[site] - self.site_theta[ds_i]
            
            
            e_u = -self.J* (np.cos(term_1u)+np.cos(term_2u)+np.cos(term_3u)+np.cos(term_4u))

            e_v = -self.J* (np.cos(term_1v)+np.cos(term_2v)+np.cos(term_3v)+np.cos(term_4v))
            
            
            delE = e_v - e_u
            ## flip the spin if the move lowers the energy
            if (delE<0.0):
                self.site_theta[site] = field[site]
            else:
                ### flip the spin with a probability
                rand = np.random.uniform()
                prob=np.exp(-delE/temp)
                if (rand<prob):
                    self.site_theta[site] = field[site]
    
    ### calculating the energy given a spin (site_theta) configuration
    def enr_calc(self):
        temp_enr = 0.0
        for i in range(self.n_sites):
            ri = self.right[i]
            li = self.left[i]
            ui = self.up[i]
            di = self.down[i]
            
            eri = np.cos(self.site_theta[i]-self.site_theta[ri])
            eli = np.cos(self.site_theta[i]-self.site_theta[li])
            eui = np.cos(self.site_theta[i]-self.site_theta[ui])
            edi = np.cos(self.site_theta[i]-self.site_theta[di])
            temp_enr += -0.5*self.J*(eri + eli + eui + edi)
        self.enr = temp_enr/self.n_sites
        
    def mag_calc(self):
        temp_magx = 0.0
        temp_magy = 0.0
        temp_magx = reduce(lambda i,j:i+j,list(map(lambda x:np.cos(x),self.site_theta)))
        temp_magy = reduce(lambda i,j:i+j,list(map(lambda x:np.sin(x),self.site_theta)))
        self.mag = np.sqrt((temp_magx)**2 + (temp_magy)**2)/self.n_sites

    ### call this function to store energy and magnetization in list
    def enr_mag_clct(self):
        self.mag_list.append(self.mag)
        self.enr_list.append(self.enr)
        
    ### spatial image of a theta configuration
    def spatial_imag(self,intp='none'):
        img = np.array(self.site_theta).reshape(self.L,self.L)
        plt.imshow(img,origin='lower',interpolation=intp,cmap='Paired')
        plt.xticks([])
        plt.yticks([])
        plt.colorbar()
        plt.show()
    
    ### time series of enrgy and magnetisation as an observable 
    def plot_timeseries(self):
        fig,ax1 = plt.subplots()
        color='tab:red'
        ax1.plot(self.enr_list,color=color)
        ax1.set_ylabel('Energy',color=color,fontsize=16)
        ax1.set_xlabel('Monte Carlo Steps',fontsize=16)
        ax1.tick_params(axis='y',labelcolor=color,labelsize=14)
        ax1.tick_params(axis='x',labelsize=14)
        ax2 = ax1.twinx()

        color='tab:green'
        ax2.plot(self.mag_list,color=color)
        ax2.set_ylabel('Magnetisation',color=color,fontsize=16)
        ax2.tick_params(axis='y',labelcolor=color,labelsize=14)

        plt.show()
    
    ### time series of only one of the observable
    def oneob_timeseries(self,enr=0):
        fig,ax1 = plt.subplots()
        color='tab:red'
        if enr==1:
            ax1.plot(self.enr_list,color=color)
            ax1.set_ylabel('Energy',color=color,fontsize=16)
            ax1.set_xlabel('Monte Carlo Steps',fontsize=16)
            ax1.tick_params(axis='y',labelcolor=color,labelsize=14)
            ax1.tick_params(axis='x',labelsize=14)
        else:
            ax1.plot(self.mag_list,color=color)
            ax1.set_ylabel('Magnetisation',color=color,fontsize=16)
            ax1.set_xlabel('Monte Carlo Steps',fontsize=16)
            ax1.tick_params(axis='y',labelcolor=color,labelsize=14)
            ax1.tick_params(axis='x',labelsize=14)
       
        plt.show()


    ### to plot the angles of the system 
    def angle_plot_arrow(self,lattfig_flag=0,wnd_flag=0):
        self.uvec = list(map(lambda x:np.cos(x),self.site_theta))
        self.vvec = list(map(lambda x:np.sin(x),self.site_theta))
        self.xind = list(map(lambda x:x%self.L,self.sites))
        self.yind = list(map(lambda x:x//self.L,self.sites))
        self.xvortex = np.array(self.xind) + 0.5
        self.yvortex = np.array(self.yind) + 0.5

        fig,ax=plt.subplots(figsize=(10,10))
        #plt.quiver(self.xind,self.yind,self.uvec,self.vvec,self.site_theta,units='xy',angles='xy'\
        #               ,scale=1.1,width=0.3,headwidth=1.3,headlength=3.5,cmap=cmap)
        plt.quiver(self.xind,self.yind,self.uvec,self.vvec,units='xy',angles='xy'\
                       ,scale=1.4,width=0.3,headwidth=1.0,headlength=4.0)
        if lattfig_flag==1:
            ims = np.array(self.site_theta).reshape(self.L,self.L)
            plt.imshow(ims,origin='lower',cmap='Paired',alpha=0.6,interpolation='bicubic')
            plt.colorbar()
        if wnd_flag==1:
            ims = np.array(self.winding_num).reshape(self.L,self.L)
#            plt.imshow(ims,origin='lower',cmap='afmhot',alpha=0.8,interpolation='bicubic')
            #plt.imshow(ims,origin='lower',cmap='RdBu',alpha=0.8)
            plt.scatter(self.xvortex,self.yvortex,c=self.winding_num)
            #plt.show()
            plt.colorbar()
        
        plt.axis([-1,self.L,-1,self.L])
        plt.xticks([])
        plt.yticks([])
        ax.spines['top'].set_visible(False)
        ax.spines['bottom'].set_visible(False)
        ax.spines['left'].set_visible(False)
        ax.spines['right'].set_visible(False)
        plt.show()

    ### this function will calculate the mod of the number applicable for both
    ### positive and negatice number.
    ### the positive number is closed at pi and negative at -pi
    def calc_modulus(self,x):
        if x<0:
            return (-1)*((abs(x))%(np.pi))
        else:
            return x%(np.pi)
    
    ### this function will first normalize theta such that they are in the range [-pi,pi]
    ### then the phase winding is calculated also taking into account that the difference can 
    ### only lie in the range.
    def windcalculation(self):
        
        for i in spin.sites:
            ri = self.right[i]
            uri = self.up[self.right[i]]
            ui = self.up[i]
            
            w21 = self.calc_modulus(self.site_theta[ri] - self.site_theta[i])
            w32 = self.calc_modulus(self.site_theta[uri] - self.site_theta[ri])
            w43 = self.calc_modulus(self.site_theta[ui] - self.site_theta[uri])
            w14 = self.calc_modulus(self.site_theta[i] - self.site_theta[ui])
            self.winding_num.append((w21+w32+w43+w14)/(2.*np.pi))
```

```python
spin=Spin()

spin.calc_n_sites()
spin.neighbour()


np.random.seed(1)

spin.theta_init()


temp = 0.3
# the new value of the theta angle that we want to change
#field_dirctn = np.random.uniform(-np.pi,np.pi,size=spin.Equil)
for j in range(spin.Equil):
    ## order at which random sites are picked for the updates
    field_dirctn = np.random.uniform(-np.pi,np.pi,size=spin.n_sites)

    rand_sites = np.random.randint(0,spin.n_sites,size=spin.n_sites)
    spin.montecarlo_sweep_single_flip(temp,field_dirctn,rand_sites)
    spin.enr_calc()
    spin.mag_calc()
    spin.enr_mag_clct()


```

```python
from sklearn.decomposition import PCA,IncrementalPCA
```

```python
### CLass declaration
class pca_class():
    def __init__(self,arr1):
        self.image = arr1
        self.var = None
        self.cumvar = None
        self.k = None
        self.recon_image = None
    
    def pca(self):
        pca = PCA()
        pca.fit(self.image)
        self.var = pca.explained_variance_ratio_*100
        self.cumvar = np.cumsum(pca.explained_variance_ratio_)*100
        self.k = np.argmax(self.cumvar>95)
        
        
    def pca_vis(self,*args):
        
        f,ax = plt.subplots(1,2,figsize=(10,4))
        ax[0].plot(self.var,marker='o',color='red',)
        ax[0].set_xlabel('# of component')
        ax[0].set_ylabel('Explained variance ratio')

        ax[1].plot(self.cumvar,marker='o',color='green')
        ax[1].axvline(x=self.k,color='red',linestyle='--')
        ax[1].axhline(y=95,color='blue',linestyle='--')
        ax[1].set_xlabel('# of component')
        ax[1].set_ylabel('Cumulative Explained variance ratio')

        ax[1].annotate("95 % variance",xy=(self.k,95),xytext=(15,55),arrowprops=dict(arrowstyle='->'))
        
        if args:
            plt.savefig(f'{args[0]}',dpi=300,bbox_inches='tight')
            plt.show()
        else:
            plt.show()
            
    def pca_inc(self,n):
        self.n_c = n
        incp = IncrementalPCA(n)
        self.recon_image = incp.inverse_transform(incp.fit_transform(self.image))
    
    def recon_vs_original_image(self,*args):
        f,ax = plt.subplots(1,2,figsize=(6,4))
        im1=ax[0].imshow(self.image,origin='lower',cmap='afmhot')#,interpolation='bilinear')
        ax[0].set_xticks([])
        ax[0].set_yticks([])
        ax[0].set_title('Original image')
        plt.colorbar(im1,ax=ax[0],shrink=0.57)
        
        im2=ax[1].imshow(self.recon_image,origin='lower',cmap='afmhot')#,interpolation='bilinear')
        ax[1].set_xticks([])
        ax[1].set_yticks([])
        ax[1].set_title(f'Reconstructed image')
        plt.colorbar(im2,ax=ax[1],shrink=0.57)
        ax[1].text(1,-4,f'{self.n_c} components',fontsize=12)
        if args:
            #plt.savefig(f'{args[0]}',dpi=300,bbox_inches='tight')
            plt.show()
        else:
            plt.show()
        
```

```python
pca1 = pca_class(arr1)
pca1.pca()
pca1.pca_vis()
pca1.pca_inc(5)
pca1.recon_vs_original_image()
```

```python
pca2 = pca_class(arr2)
pca2.pca()
pca2.pca_vis()
pca2.pca_inc(11)
pca2.recon_vs_original_image()
```
